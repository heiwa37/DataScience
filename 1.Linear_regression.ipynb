```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as mp
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/Users/rrashmi/ML/original-1.csv')
```

```python
data.describe()
```

```python
#printing  data
```

```python
X= DataFrame(data,columns=['production_budget_usd'])
Y= DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
mp.figure(figsize=(10,6))
mp.scatter(X,Y,alpha=0.5)
mp.title("film vs movie budget")
mp.xlabel('Budget')
mp.xlim(0,450000000)
mp.ylim(0,3000000000)
mp.ylabel('Revenue')
```

'Linear Regression'

```python
regression = LinearRegression()
regression.fit(X,Y) #creating a curve on the graph
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
mp.figure(figsize=(10,6))
mp.scatter(X,Y,alpha=0.5)
mp.title("film vs movie budget with Slope overview")
mp.xlabel('Budget')
mp.xlim(0,450000000)
mp.ylim(0,3000000000)
mp.ylabel('Revenue')
mp.plot(X,regression.predict(X),color='Black', Linewidth=4)
```

SO we got the slope, m and c value from the algorithm, 
1. with this info , feed input  say x=50 m, what is my target= h(x)

```python
regression.score(X,Y) # this gives the probability of model
                    # considering our values
```
