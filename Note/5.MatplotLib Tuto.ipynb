```python
np.pi
```

```python
x= np.arange(0,5*np.pi,0.1)
plt.subplot(2,1,1)
y=np.sin(x) # calling sine function
plt.plot(x,y,'b--')  # plotting that sine function
plt.title("sine wave")
# plotting cos function
plt.subplot(2,1,2)
z=np.cos(x)
plt.plot(x,z,'r')
plt.title('cosine wave')
```

```python
# Bar Graph
x=[2,6,10]
y=[6,8,12]

x2=[1,3,5]
y2=[3,5,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
```

## Box Plot- helps to find percentile ( stats)

```python
data=[np. random.normal(0,std,100)for std in range(1,4)] # normal distribution
data # type: list
```

```python
plt.boxplot(data,vert=True,patch_artist=False); # ; is important while plotting
```

```python
#pie chart
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True) #%1d for int type

plt.axis('equal')
plt.show()
```
