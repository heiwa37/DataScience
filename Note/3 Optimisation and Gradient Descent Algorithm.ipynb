# Optimisation and Gradient Descent Algorithm

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Example 1- A simple cost function

$$f(x)= x^2+x+1 $$

```python
def f(x):
    return x**2+x +1
```

```python
#make data 
x_1=np.linspace(start=-3, stop=3, num =100) #100 value of range from -3 to 3 (array)
x_1
```

```python
#plot
plt.xlim(-3,3)
plt.ylim(0,8,1)
plt.xlabel('x',fontsize=12)
plt.ylabel('f(x)',fontsize=12)
plt.plot(x_1,f(x_1))
plt.show()
```

## Slope & Derivates

```python
#slope of x= derivate of x
def df(x):
    return (2*x+1)
```

```python
#plot
#1 chart: cost of the function
plt.figure(figsize=[15,5])
plt.subplot(1,2,1)
plt.xlim(-3,3)
plt.ylim(0,8,1)
plt.xlabel('x',fontsize=12)
plt.ylabel('f(x)',fontsize=12)
plt.plot(x_1,f(x_1))
plt.show()

#2. Seconfd chart:  slope of cost function
plt.figure(figsize=[15,5])
plt.subplot(1,2,2)
plt.xlim(-2,3)
plt.ylim(-3,8)
plt.xlabel('x',fontsize=12)
plt.ylabel('df(x)',fontsize=12)
plt.plot(x_1,df(x_1),color='r',linewidth=2)
plt.grid()
plt.show()
```

```python
#Gradient Descent
new_x=3 # starting point
previous_x= 0
step_multiplier=0.1
for n in range(500):
        previous_x=new_x
        gradient=df(previous_x) # c
        new_x=previous_x- step_multiplier*gradient
print('Local minimum occurs at:', new_x)
print('slope of the point is :', df(new_x))
print('f(x) value of cost fucntion is :',f(new_x))


        
        
```
